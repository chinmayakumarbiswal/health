```python
import pandas as pd
dataset=pd.read_excel('health_care1.xlsx')
print(dataset)
```

```python
import plotly.graph_objects as go
fig = go.Figure(go.Bar(x=dataset.age,y=dataset.heartdisease,text=dataset.heartdisease,textposition='outside'))
fig.update_layout(title="Total heartdisease patient wrt age",xaxis_title="Age",yaxis_title="heartdisease patient")
fig.update_traces(textfont_size=10,textfont_color='red')
fig.show()
```

```python
import plotly.graph_objects as go
fig = go.Figure(go.Bar(x=dataset.age,y=dataset.hypertension,text=dataset.hypertension,textposition='outside'))
fig.update_layout(title="Total hypertension patient wrt age",xaxis_title="Age",yaxis_title="Total hypertension patient")
fig.update_traces(textfont_size=10,textfont_color='black')
fig.show()
```
